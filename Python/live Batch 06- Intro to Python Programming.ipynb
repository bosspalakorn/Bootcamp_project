# OOP
Object Oriented Programing 

```python
class Dog:
    def __init__(self,name,age,color):
       self.name  = name 
       self.age   = age 
       self.color = color
        
# function (Dog method)
    def hello(self):
        print(f"Hi my name is {self.name}!")

    def get_older(self):
        self.age +=1
        print(f"I'm getting older one year. I'm now {self.age} years old. ")
```

```python
dog1 = Dog("milo" ,2,"black")
dog2 = Dog("rambo",3,"golden")

dog1.hello()
dog2.hello()
```

```python
dog1
```

```python
dog1.get_older()
```

```python
dog2.get_older()
```

```python
print(dog1.name ,dog1.age,dog1.color)
```

```python
class ATM:
     # double underscore => dunder
     def __init__(self,name,balance):
         self.name = name 
         self.balance = balance

     def check_balance(self):
         message = f"Account:{self.name},Balance:{self.balance}"
         print(message)
     def deposit (self,money):
         self.balance += money
         print (f"New Balance: {self.balance}")
         print("Deposit successfully.")
     def change_name(self,new_acc_name):
         self.name = new_acc_name
         print(f"New Name:{self.name}")
         print("Your account name has been changed.")
```

```python
scb = ATM("Anna", 300)
```

```python
print(scb.name,scb.balance)
```

```python
scb.check_balance()
```

```python
scb.deposit(900)
```

```python
scb.balance
```

```python
scb.name
```

```python
scb.change_name("Marry Anne")
```

```python
scb.name
```
